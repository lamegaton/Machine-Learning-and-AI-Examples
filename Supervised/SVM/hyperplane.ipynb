```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 2D Hyperplane example setup
# why do we need this?
# because we need to input x (data points) from -10 to 10
x_vals_2d = np.linspace(-5, 5, 10)

# Example 1: w = [1, 1], b = -2 (2D plane)
y_vals_2d_1 = 2 - x_vals_2d  # Equation: x1 + x2 - 2 = 0 => x2 = 2 - x1

# Example 2: w = [1, 0], b = -3 (Vertical line x1 = 3)
y_vals_2d_2 = np.full(x_vals_2d.shape, np.nan)  # No slope, vertical line

# Example 3: w = [0, 1], b = 4 (Horizontal line x2 = -4)
y_vals_2d_3 = np.full(x_vals_2d.shape, -4)  # Horizontal line at x2 = -4

# Plot 2D hyperplanes
plt.figure(figsize=(10, 5))
plt.plot(x_vals_2d, y_vals_2d_1, label="w=[1,1], b=-2", color='b')
plt.axvline(x=3, color='g', linestyle='--', label="w=[1,0], b=-3")
plt.plot(x_vals_2d, y_vals_2d_3, label="w=[0,1], b=4", color='r')
plt.title('2D Hyperplane Examples')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)
plt.legend()

# 3D Hyperplane example setup
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Define the grid for 3D plot
x_vals_3d, y_vals_3d = np.meshgrid(x_vals_2d, x_vals_2d)

# Example 4: w = [2, 3, -1], b = -5 (3D hyperplane)
z_vals_3d_4 = (5 - 2 * x_vals_3d - 3 * y_vals_3d)  # Equation: 2x1 + 3x2 - z - 5 = 0 => z = (5 - 2x1 - 3x2)

# Plot the 3D hyperplane
ax.plot_surface(x_vals_3d, y_vals_3d, z_vals_3d_4, color='c', alpha=0.7)
ax.set_title('3D Hyperplane Example: w=[2,3,-1], b=-5')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3 (z)')

plt.show()
```
